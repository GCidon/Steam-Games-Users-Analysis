```python
import csv
import pandas
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pandas.read_csv('maxPricePrecio-Developer.csv')

plt.style.use('ggplot')

x = df[df.columns[0]]
y = df[df.columns[1]]
x_pos = [i for i, _ in enumerate(x)]
plt.barh(x_pos, y, color='green')
plt.ylabel("Developer")
plt.xlabel("Average price of developer's software")
plt.title("Top 10 Steam genres by time users spend on them")
plt.yticks(x_pos, x)

plt.show()
```
